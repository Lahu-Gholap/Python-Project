## 2. Univariate Analysis –

### 1.Plot the distribution of movie release years (Histogram)

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['ReleaseYear'], bins=20, kde=False)
plt.title("Distribution of Movie Release Years")
plt.xlabel("ReleaseYear")
plt.ylabel("Count of Movies")
plt.show()
```

### 2. Top 5 most common genres (Bar Chart) 

```python
top_genres = df['Genre'].value_counts().head(5)

plt.figure(figsize=(8,5))
sns.barplot(x=top_genres.index, y=top_genres.values)
plt.title("Top 5 Most Common Genres")
plt.xlabel("Genre")
plt.ylabel("Number of Movies")
plt.show()
```

### 3.Plot the distribution of IMDb ratings (Histogram / KDE)

```python
plt.figure(figsize=(8,5))
sns.histplot(df['IMDbRating'], kde=True, bins=20)
plt.title("Distribution of IMDbRating")
plt.xlabel("IMDbRating")
plt.ylabel("Frequency")
plt.show()
```

### 4.Show the distribution of Rotten Tomatoes scores (Histogram)

```python
plt.figure(figsize=(8,5))
sns.histplot(df['RottenTomatoesScore'], bins=20, kde=False)
plt.title("Distribution of Rotten Tomatoes Scores")
plt.xlabel("Rotten Tomatoes Score (%)")
plt.ylabel("Count of Movies")
plt.show()
```

### 5. Average budget of movies by country (Bar chart)

```python
avg_budget_country = df.groupby('Country')['BudgetUSD'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,5))
sns.barplot(x=avg_budget_country.index, y=avg_budget_country.values)
plt.xticks(rotation=45)
plt.title("Average Movie Budget by Country")
plt.xlabel("Country")
plt.ylabel("Average Budget")
plt.show()
```

## 3. Bivariate Analysis


### 1.Correlation: Budget vs Global Box Office (Scatter Plot)

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['BudgetUSD'], y=df['Global_BoxOfficeUSD'])
plt.title("Budget vs Global Box Office Revenue")
plt.xlabel("Budget")
plt.ylabel("Global_BoxOfficeUSD")
plt.show()
corr = df['BudgetUSD'].corr(df['Global_BoxOfficeUSD'])
print("Correlation between Budget and Global Revenue:", corr)
```

 ### 2. Average IMDb Rating Across Genres (Bar Plot & Box Plot)

```python
# BAR plot
avg_imdb = df.groupby('Genre')['IMDbRating'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,5))
sns.barplot(x=avg_imdb.index, y=avg_imdb.values)
plt.title("Average IMDb Rating by Genre")
plt.xticks(rotation=45)
plt.ylabel("Average IMDb Rating")
plt.show()
```

```python
#BOX plot
plt.figure(figsize=(10,5))
sns.boxplot(x=df['Genre'], y=df['IMDbRating'])
plt.title("Distribution of IMDb Ratings by Genre")
plt.xticks(rotation=45)
plt.show()
```

### 3. Do higher budgets lead to better Rotten Tomatoes scores? (Scatter Plot)

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['BudgetUSD'], y=df['RottenTomatoesScore'])
plt.title("Budget vs Rotten Tomatoes Score")
plt.xlabel("Budget")
plt.ylabel("Rotten Tomatoes Score")
plt.show()

corr_rt = df['BudgetUSD'].corr(df['RottenTomatoesScore'])
print("Correlation between Budget and Rotten Tomatoes:", corr_rt)
```

### 4. US vs Global Revenue by Genre (Which genres depend on international sales?)

```python
rev_by_genre = df.groupby('Genre')[['US_BoxOfficeUSD', 'Global_BoxOfficeUSD']].mean()

rev_by_genre.plot(kind='bar', figsize=(10,5))
plt.title("US vs Global Revenue by Genre")
plt.xlabel("Genre")
plt.ylabel("Average Revenue")
plt.xticks(rotation=45)
plt.show()
```

### 5. Relationship between Opening Day Sales and First Week Sales

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['Opening_Day_SalesUSD'], y=df['One_Week_SalesUSD'])
plt.title("Opening Day vs First Week Sales")
plt.xlabel("Opening Day Sales")
plt.ylabel("First Week Sales")
plt.show()
corr_sales = df['Opening_Day_SalesUSD'].corr(df['One_Week_SalesUSD'])
print("Correlation:", corr_sales)
```

## 4. Time-Based Analysis

### 1. How has the number of movies released per year changed over time? (Line chart).


```python
df['Release Year'] = pd.to_numeric(df['ReleaseYear'], errors='coerce')

movies_per_year = df.groupby('ReleaseYear').size()

plt.figure(figsize=(10,4))
plt.plot(movies_per_year.index, movies_per_year.values)
plt.title("Number of Movies Released per Year")
plt.xlabel("Year")
plt.ylabel("Number of Movies")
plt.grid(True)
plt.show()
```

### 2. Has the average movie budget increased over the decades? (Line/Bar chart).

```python
#Line plot
df['Decade'] = (df['Release Year'] // 10) * 10

avg_budget_decade = df.groupby('Decade')['BudgetUSD'].mean()

plt.figure(figsize=(10,4))
plt.plot(avg_budget_decade.index, avg_budget_decade.values, marker='o')
plt.title("Average Movie Budget by Decade")
plt.xlabel("Decade")
plt.ylabel("Average Budget")
plt.grid(True)
plt.show()
```

```python
#bar plot
plt.figure(figsize=(10,4))
plt.bar(avg_budget_decade.index, avg_budget_decade.values)
plt.title("Average Movie Budget by Decade")
plt.xlabel("Decade")
plt.ylabel("Average Budget")
plt.show()
```

### 3. Do IMDb ratings show any trend across release years? (Line chart)

```python
avg_imdb_year = df.groupby('ReleaseYear')['IMDbRating'].mean()

plt.figure(figsize=(10,4))
plt.plot(avg_imdb_year.index, avg_imdb_year.values, color='green')
plt.title("Average IMDb Rating by Release Year")
plt.xlabel("Year")
plt.ylabel("IMDb Rating")
plt.grid(True)
plt.show()
```

### 4. Which genre has grown most in popularity since 2000? (Line/Bar chart)

```python
df_after_2000 = df[df["ReleaseYear"] >= 2000]
genre_count = df_after_2000["Genre"].value_counts()

plt.figure(figsize=(10,5))
plt.bar(genre_count.index, genre_count.values)
plt.xticks(rotation=70)
plt.xlabel("Genre")
plt.ylabel("Number of Movies")
plt.title("Genre Popularity Since 2000")
plt.show()
```

### 4.Compare average Global Box Office revenue of movies before 2000 vs after 2000 (Bar chart).

```python
df["Period"] = df["ReleaseYear"].apply(lambda x: "Before 2000" if x < 2000 else "After 2000")
avg_revenue = df.groupby("Period")["Global_BoxOfficeUSD"].mean()

plt.figure(figsize=(7,5))
plt.bar(avg_revenue.index, avg_revenue.values)
plt.xlabel("Period")
plt.ylabel("Avg Global Box Office")
plt.title("Global Box Office Revenue Before vs After 2000")
plt.show()
```

## 5. Categorical Insights


### 1.Which director has the highest average IMDb rating? (Top 10 directors bar chart)

```python
import matplotlib.pyplot as plt
import pandas as pd

# Top 10 directors based on average IMDb rating
director_rating = (
    df.groupby("Director")["IMDbRating"].mean()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(10,5))
plt.bar(director_rating.index, director_rating.values)
plt.xticks(rotation=70)
plt.ylabel("Average IMDb Rating")
plt.title("Top 10 Directors by IMDb Rating")
plt.show()
```

## 2.Which lead actors have appeared in the most movies? (Bar chart).

```python
# Count movies by lead actor
actor_count = df["LeadActor"].value_counts().head(10)

plt.figure(figsize=(10,5))
plt.bar(actor_count.index, actor_count.values)
plt.xticks(rotation=70)
plt.ylabel("Number of Movies")
plt.title("Top Lead Actors by Movie Count")
plt.show()
```

### 3.Which country produces the most movies? (Bar chart).

```python
# Count movies by country
country_count = df["Country"].value_counts().head(10)

plt.figure(figsize=(10,5))
plt.bar(country_count.index, country_count.values)
plt.xticks(rotation=70)
plt.ylabel("Number of Movies")
plt.title("Top Movie-Producing Countries")
plt.show()
```

### 4.Which genre tends to have the highest US vs Global box office ratio?

```python
# Calculate ratio of US Box Office to Global Box Office for each movie
df["US_to_Global_Ratio"] = df["US_BoxOfficeUSD"] / df["Global_BoxOfficeUSD"]

# Average ratio by genre
genre_ratio = (
    df.groupby("Genre")["US_to_Global_Ratio"].mean()
      .sort_values(ascending=False)
)

plt.figure(figsize=(10,5))
plt.bar(genre_ratio.index, genre_ratio.values)
plt.xticks(rotation=20)
plt.ylabel("Avg US / Global Box Office Ratio")
plt.title("Genre with Highest US vs Global Box Office Ratio")
plt.show()
```

### 5.Compare IMDb and Rotten Tomatoes ratings for each genre — are they aligned? 

```python
import numpy as np
```

```python

# Group by Genre and compute mean IMDb & Rotten Tomatoes ratings
genre_compare = df.groupby("Genre")[["IMDbRating", "RottenTomatoesScore"]].mean()

# Prepare values
genres = genre_compare.index
imdb = genre_compare["IMDbRating"]
rt = genre_compare["RottenTomatoesScore"]

# Set bar positions
x = np.arange(len(genres))
width = 0.35  # width of each bar

plt.figure(figsize=(12,6))

# Side-by-side bars
plt.bar(x - width/2, imdb, width, label="IMDb Rating")
plt.bar(x + width/2, rt, width, label="Rotten Tomatoes")

# Labels & details
plt.xticks(x, genres, rotation=70)
plt.ylabel("Average Rating")
plt.title("IMDb vs Rotten Tomatoes Ratings by Genre")
plt.legend()

plt.tight_layout()
plt.show()
```

## 6.Visualization Ideas

### 1.Heatmap of correlations among numeric variables (Budget, Revenue, Ratings, Votes).

```python
numeric_cols = ["BudgetUSD", "Global_BoxOfficeUSD", "IMDbRating", "RottenTomatoesScore", "NumVotesIMDb"]

corr = df[numeric_cols].corr()

plt.figure(figsize=(10,5))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Numeric Variables")
plt.show()
```

### 2.Bubble chart of Budget vs Global Revenue, with bubble size = IMDb votes, color = Genre

```python
plt.figure(figsize=(10,6))
scatter = plt.scatter(
    df["BudgetUSD"],
    df["Global_BoxOfficeUSD"],
    s=df["NumVotesIMDb"] / 1000,   # Scale bubble size
    c=df["Genre"].astype('category').cat.codes,  # Color by Genre
    alpha=0.6
)

plt.xlabel("Budget")
plt.ylabel("Global Box Office")
plt.title("Budget vs Global Revenue (Bubble = IMDb Votes, Color = Genre)")
plt.colorbar(scatter, label="Genre Code")
plt.grid(True)
plt.show()
```

### 3.Trend line of IMDb rating vs Release Year.

```python
df_year = df.groupby("ReleaseYear")["IMDbRating"].mean()

plt.figure(figsize=(10,5))
plt.plot(df_year.index, df_year.values, marker='o')
plt.xlabel("ReleaseYear")
plt.ylabel("Average IMDb Rating")
plt.title("Trend of IMDb Rating Over Time")
plt.grid(True)
plt.show()
```

### 4.Pie chart of genre distribution

```python
genre_counts = df["Genre"].value_counts()

plt.figure(figsize=(8,8))
plt.pie(genre_counts.values, labels=genre_counts.index, autopct="%1.1f%%")
plt.title("Genre Distribution")
plt.show()
```

### 5.Box plot of IMDb rating by Genre.

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x="Genre", y="IMDbRating")
plt.xticks(rotation=70)
plt.title("Box Plot: IMDb Rating by Genre")
plt.show()
```
